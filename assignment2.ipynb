```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

# Loading csv file

```python
df = pd.read_csv('/content/House Price India.csv')
df
```

# Univariate Analysis

```python
#pie plot
df['Number of schools nearby'].value_counts()
plt.pie(df['Number of schools nearby'].value_counts(), labels = df['Number of schools nearby'].unique(), autopct='%0.01f%%')
plt.show()
```

```python
#dist plot
sns.distplot(df['number of floors'])
```

```python
#bar plot
sns.barplot(x=df['number of bedrooms'].value_counts().index,y=df['number of bedrooms'].value_counts())
```

# Bivariate Analysis

```python
#Line Plot
sns.lineplot(x=df['number of bedrooms'],y=df['number of bathrooms'])
```

```python
#Scatterplot
sns.scatterplot(x=df['number of floors'],y=df['living area'])
```

# Multivariate Analysis

```python
plt.figure(figsize=(16,24))
#HeatMap
sns.heatmap(df.corr(),annot=True)
```

```python
#PairPlot
sns.pairplot(df)
```
